````python
import json

def normalize_and_save_data(input_file, output_file):
    # Đọc dữ liệu từ tệp JSON
    with open(input_file, 'r', encoding='utf-8') as f:
        data = json.load(f)
    
    normalized_list = []
    for entry in data:
        # Lấy nội dung JSON từ trường "content"
        json_str = entry["content"].strip("```json\n").strip("```")
        try:
            # Chuyển đổi chuỗi JSON thành từ điển
            json_data = json.loads(json_str)
            normalized_list.append(json_data)
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON for entry: {entry}, Error: {e}")
    
    # Lưu danh sách normalized_list vào file JSON mới
    with open(output_file, 'w', encoding='utf-8') as f:
        json.dump(normalized_list, f, ensure_ascii=False, indent=4)

# Ví dụ sử dụng
input_file = 'messages.json'  # Tên tệp JSON đầu vào
output_file = 'output.json'  # Tên tệp JSON đầu ra

normalize_and_save_data(input_file, output_file)
print("Data has been normalized and saved to output.json.")
````
